```python
#Importing Required Libraries
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ravi-raj-97/ml_grad_desc/master/airfoil_self_noise.dat', sep='\t',header = None)
# Columns of the data as per the website where the repository has been downloaded.
data.columns = ['Frequency','angle_of_attack','chord_length','free_stream_velocity','suction_side_displacement_thickness','scaled_sound_pressure_level']
data.head(10)
```

```python
#percentage of missing values
data.isnull().sum()*100/len(data)
```

Dataset is clean with no missing data

```python
#size of data
data.shape
```

```python
#description of each column
data.describe()
```

```python
#data visualisation
sns.displot(data['scaled_sound_pressure_level'],kde = True, bins=20)
plt.show()
```

From the above plot it can be approximated that target variable (*scaled_sound_pressure_level*) is normally distributed.

```python
correlation_matrix = data.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

All the attributes can be used as features for target variable
